```python
# fix the dictionary

# need space in the beginning of dictionary

# 0th doesnt exist in our wordindex, but 1,2,3 do!

# begin by shifting all the words by 3 positions!!!

new_dictionary = {}
for word,index in wordindex.items():
  new_dictionary[word] = index + 3

# space has been made to insert the NLP flags of pad, start, unknown and unused

new_dictionary["<PAD>"] = 0
new_dictionary["<START>"] = 1
new_dictionary["<UNK>"] = 2
new_dictionary["<UNUSED>"] = 3
```

```python
rev_new_dict = {}
for word,index in new_dictionary.items():
  rev_new_dict[index] = word
```

```python
print(rev_new_dict)
```

```python
def new_decoder(input):
  decoded_sentence = []
  for inp in input:
    decoded_sentence.append(rev_new_dict[inp])
  return decoded_sentence
```

```python
HP_WORDS_ALLOWED = 10000
```

```python
#just for demo, let's consider a subset of these 88,588 words!
# maybe top-10K!

# <UNK>
(xtrain,ytrain),(xtest, ytest) = imdb.load_data(num_words=HP_WORDS_ALLOWED)
```

```python
# ytrain, ytest should contain 1s and 0
ytrain[:2], ytest[:2]
```

```python
for i in range(100,105):
  print()
  print(" ".join(new_decoder(xtrain[i])))
```

```python
model = keras.Sequential([
    keras.layers.Embedding(input_dim=HP_WORDS_ALLOWED,output_dim=32 )
])
```

```python
model.summary() # empty embedding layer
```

```python
# prompt: pass a sentence to my model with embedding layer only and show the output vectors

# Example sentence
sentence = xtrain[0]

# Convert the sentence to a tensor
sentence_tensor = tf.convert_to_tensor(sentence)

# Add a batch dimension
sentence_tensor = tf.expand_dims(sentence_tensor, axis=0)

# Pass the sentence to the model
output = model(sentence_tensor)

# Print the output vectors
print(output)
```
